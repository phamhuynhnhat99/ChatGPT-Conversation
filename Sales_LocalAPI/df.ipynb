```python
import pandas as pd
from sklearn.metrics import confusion_matrix, classification_report
import random
```

```python
path_ = '/home/aia/Nhat/ChatGPT-Conversation/Sales_LocalAPI/gpt lan 3/predict_'
df = pd.read_csv('/home/aia/Nhat/ChatGPT-Conversation/Sales_LocalAPI/gpt lan 3/predict_xxx.csv', dtype=str, keep_default_na=False)
for i in range(10):
    j = 9-i
    name_ = path_ + str(j)+ ".csv"
    tmp_ = pd.read_csv(name_, dtype=str, keep_default_na=False)
    df = pd.concat([tmp_, df], axis=0)
label_names = list(range(0, 7+1))
print(df.shape)
df.head()
```

```python
y_true_str = [str(_) for _ in df.y_true] + label_names
y_pred_str = [str(_) for _ in df.y_pred] + label_names
clf_rp = classification_report(y_true_str, y_pred_str, target_names = ["0", "1", "2", "3", "4", "5", "6", "7"])
print(clf_rp)
cm = confusion_matrix(y_true_str, y_pred_str)
cm_df = pd.DataFrame(cm)
cm_df.columns = label_names
name = "y_true"
cm_df[name] = label_names
cm_df.set_index(name, inplace=True)
cm_df
```
